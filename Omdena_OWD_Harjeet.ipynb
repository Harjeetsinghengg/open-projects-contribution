```python
import pandas as pd

# URL of the CSV file
url = "https://drive.google.com/uc?id=1QXyicFZy8_TIWQRjI80o8isxCp-CX_Mr"

# Load the CSV file into a DataFrame
df = pd.read_csv(url)

# Filter data for Kenya
kenya_data = df.loc[df['country'] == 'Kenya']

# Display the first few rows of the filtered DataFrame
print(kenya_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = kenya_data

# Filter data for Kenya
kenya_data = data[data['country'] == 'Kenya']

# 1. Data Cleaning
# Check for missing values
missing_values = kenya_data.isnull().sum()
print("Missing Values:\n", missing_values)

# 2. Summary Statistics
summary_stats = kenya_data.describe()
print("Summary Statistics:\n", summary_stats)

# 3. Data Visualization
# Histograms for numerical variables
numerical_vars = ['population', 'gdp', 'cement_co2', 'co2', 'temperature_change_from_co2']
kenya_data[numerical_vars].hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Bar chart for categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x='iso_code', data=kenya_data)
plt.title('ISO Code Distribution for Kenya')
plt.xlabel('ISO Code')
plt.ylabel('Frequency')
plt.show()

# Box plot for numerical variables
plt.figure(figsize=(10, 6))
sns.boxplot(data=kenya_data[numerical_vars])
plt.title('Box Plot for Numerical Variables')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()

# 4. Trends and Patterns
# Line plot for CO2 emissions over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='co2', data=kenya_data)
plt.title('CO2 Emissions in Kenya over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=45)
plt.show()

```
